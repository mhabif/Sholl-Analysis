```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
#Import data-sets
data_Sholl_wt = pd.read_csv (r'Sholl_wt.txt')
data_Sholl_Tg = pd.read_csv (r'Sholl_Tg.txt')
```

```python
# Define X values
x = data_Sholl_Tg.Distance_from_0
x1 = data_Sholl_wt.Distance_from_0

# Plot 

plt.style.use('seaborn-white') # ver https://python-graph-gallery.com/199-matplotlib-style-sheets/

plt.subplots(figsize=(18, 5))

plt.subplot(131)
plt.plot(x, data_Sholl_Tg.MEAN, linewidth=2, color='r', label='Mean ± SD (Tg, N=107)')
plt.fill_between(x, data_Sholl_Tg.MEAN, data_Sholl_Tg.MEAN + data_Sholl_Tg.STD, color='r', alpha=0.2)
plt.fill_between(x, data_Sholl_Tg.MEAN, data_Sholl_Tg.MEAN - data_Sholl_Tg.STD, color='r', alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,2)
plt.legend (loc="upper right",fontsize=13)
plt.xlabel('Radial distance from \n cell body (AU)', fontsize=14, fontweight='bold')
plt.ylabel('# Intersections', fontsize=14,fontweight='bold')

plt.subplot(132)
plt.plot(x1, data_Sholl_wt.MEAN, linewidth=2, color='b', label='Mean ± SD (wt, N=112)')
plt.fill_between(x1, data_Sholl_wt.MEAN, data_Sholl_wt.MEAN + data_Sholl_wt.SD, color='b', alpha=0.2)
plt.fill_between(x1, data_Sholl_wt.MEAN, data_Sholl_wt.MEAN - data_Sholl_wt.SD, color='b', alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,2)
plt.legend (loc="upper right",fontsize=13)
plt.xlabel('Radial distance from \n cell body (AU)', fontsize=14, fontweight='bold')
plt.ylabel('# Intersections', fontsize=14,fontweight='bold')

plt.style.use('seaborn-dark') # ver https://python-graph-gallery.com/199-matplotlib-style-sheets/

plt.subplot(133)
plt.plot(x, data_Sholl_Tg.MEAN, linewidth=2, color='r', label='Mean (Tg, N=107)')
plt.plot(x1, data_Sholl_wt.MEAN, linewidth=2, color='b', label='Mean (wt, N=112)')
plt.ylim(0,10)
plt.xlim(0,2)
plt.legend (loc="upper right", fontsize=13)
plt.xlabel('Radial distance from \n cell body (AU)', fontsize=14, fontweight='bold')
plt.ylabel('# Intersections', fontsize=14,fontweight='bold')


plt.savefig("Sholl_wt_Tg_male_13mo.png", format="png")

plt.show()
```
